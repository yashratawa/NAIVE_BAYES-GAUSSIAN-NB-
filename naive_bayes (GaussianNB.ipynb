```python
import pandas as pd
```

```python
a = pd.read_csv('train.csv')
a
```

```python
a.drop(['Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis = 'columns', inplace = True)
a
```

```python
dummy = pd.get_dummies(a.Sex)
dummy
```

```python
x = a.drop(['Survived'], axis = 'columns')
x.head()
```

```python
y = a.Survived
y.head()
```

```python
newx = pd.concat([x,dummy], axis = 'columns')
newx.head()
```

```python
newx.drop(['Sex'], axis = 'columns', inplace = True)
newx.head()
```

```python
newx.columns[newx.isna().any()] 
```

```python
newx.Age[:10]
```

```python
mean = newx.Age.mean()
mean
```

```python
newx.Age = newx.Age.fillna(30)
```

```python
newx.Age[:10]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_tr,x_ts,y_tr,y_ts = train_test_split(newx,y, test_size = 0.7)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(x_tr,y_tr)
```

```python
model.score(x_ts,y_ts)
```

```python
y_ts[:10]
```

```python
e = model.predict(x_ts.head())
e[:5]
```

```python
import matplotlib.pyplot as plt
```

```python
z={'srvived': e, "Age" : x_ts.Age, 'female' : x_ts.female, 'male' : x_ts.male, 'Passenger Id' : x_ts.PassengerId}
b=pd.DataFrame(z)
b
#this shows that according to the prediction, only female passengers survived...(1=suvived,
#                                                                                0= not survived)
```
